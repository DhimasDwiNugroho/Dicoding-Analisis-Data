### Cleaning Data

```python
#pada projek ini hanya menggunakan data day yang disesuaikan dengan pertanyaan bisnis yang telah ditentukan
#Tidak ada yang perlu dilakukan dikarenakan tidak terdapat data kosong dan duplikasi data
```

```python
#menghapus kolom instan dan windspeed
df1.drop(['instant', 'windspeed'],axis = 1, inplace = True)
```

```python
df1.head()
```

Mengubah nama kolom agar mudah dipahami

```python
df1.rename(columns={'dteday': 'date',
                    'yr': 'year',
                    'mnth': 'month',
                    'cnt': 'count',
                    'weathersit' : 'weather'}, inplace=True)
# df1.rename(columns={'yr': 'year'}, inplace=True)
# df1.rename(columns={'mnth': 'month'}, inplace=True)
# df1.rename(columns={'cnt': 'count'}, inplace=True)
df1.head()
```

Mengubah type data pada kolom

```python
df1.info()
```

```python
df1['date'] = pd.to_datetime(df1.date)
df1 = df1.astype({'year': 'category'})
df1 = df1.astype({'season': 'category'})
df1 = df1.astype({'month': 'category'})
df1 = df1.astype({'weather': 'category'})
df1 = df1.astype({'workingday': 'category'})
df1 = df1.astype({'weekday': 'category'})
```

```python
df1.info()
```

```python
print ("nilai unik pada kolom season : ", df1["season"].unique())
print ("nilai unik pada kolom year : ", df1["year"].unique())
print ("nilai unik pada kolom month : ", df1["month"].unique())
print ("nilai unik pada kolom working day : ", df1["workingday"].unique())
print ("nilai unik pada kolom weather : ", df1["weather"].unique())
print ("nilai unik pada kolom weekday : ", df1["weekday"].unique())
```

```python
df1['season'] = df1['season'].map({
    1: "springer", 2 : "summer", 3 : "fall", 4 : "winter"})

df1['year'] = df1['year'].map({
    0: "2011", 1 : "2012"})

df1['month'] = df1['month'].map({
    1: 'Jan', 2: 'Feb', 3: 'Mar', 4: 'Apr', 5: 'May', 6: 'Jun',
    7: 'Jul', 8: 'Aug', 9: 'Sep', 10: 'Oct', 11: 'Nov', 12: 'Dec'})

df1['workingday'] = df1['workingday'].map({
    0 : "holiday" , 1 : "workingday"})

df1['weather'] = df1['weather'].map({
    1 : "Clear" , 2 : "Mist or Cloudy", 3 : 'Light Snow or Light Rain', 4 : 'Heavy Rain'})

df1['weekday'] = df1['weekday'].map({
    0: 'Sun', 1: 'Mon', 2: 'Tue', 3: 'Wed', 4: 'Thu', 5: 'Fri', 6: 'Sat'})
```
